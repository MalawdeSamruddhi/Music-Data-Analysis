```python
sns.displot(df['loudness'],color='c')
```

```python
x = df['loudness']
sns.set_theme(style='whitegrid')
sns.boxplot(x,palette='Set2_r')
```

```python
sns.displot(df['energy'],color='c')
```

```python
y = df['energy']
sns.set_theme(style='whitegrid')
sns.boxplot(y,palette='Set3_r')
```

```python
plt.figure(figsize=(10,5))
x = df['loudness']
y = df['energy']
sns.scatterplot(x,y,marker="*",color='g')
```

```python
sns.jointplot(x='loudness',y='energy',data=df,kind='hist',color='m')
```

```python
x = df['duration_in min/ms']
y = df['energy']
sns.scatterplot(x,y,color='b')
```

```python
x = df['valence']
y = df['energy']
sns.relplot(x,y,kind='line',data=df,color='g')
```

```python
x = df['tempo']
y = df['energy']
sns.relplot(x,y,data=df,kind='line',color='c')
```

```python
x = df['liveness']
y = df['energy']
sns.relplot(x,y,data=df,kind='line',color='m')
```

```python
x = df['valence']
y = df['loudness']
sns.relplot(x,y,data=df,color='y')
```

CONCLUSION : The features energy and loudness are highly correlated with each other. The graph shows an exponential increase in loudness as energy goes on increasing. Energy of the music mostly depends on factors duration in mins, valence, tempo and liveness. The feature loudness is dependable on valence.
